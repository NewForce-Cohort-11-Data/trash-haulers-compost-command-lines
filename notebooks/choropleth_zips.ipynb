```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
zipcodes = gpd.read_file('../data/zipcodes.geojson')
print(zipcodes.crs)
zipcodes.info()
```

```python
zipcodes
```

```python
zipcodes.plot()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
zipcodes.plot(ax = ax)
ax.axis('off');
```

```python
fines = pd.read_csv('../data/fines.csv')

fines.info()
```

```python
fines_lean = fines.drop(columns = ['Incident Address', 'Trash Hauler', 'Trash Route', 'State Plan X', 'State Plan Y', 'Council District'])
fines_lean['Zip Code'] = fines_lean['Zip Code'].astype('Int64').astype(object)
fines_lean
```

```python
fines_by_zip = fines_lean.groupby('Zip Code').sum().sort_values('Fines', ascending=False).reset_index()
```

```python
fines_by_zip['Zip Code'] = fines_by_zip['Zip Code'].astype('object')
fines_by_zip
```

```python
fines_by_zip.columns = fines_by_zip.columns.str.strip()
```

```python
zipcodes.columns = zipcodes.columns.str.strip()
```

```python
print(zipcodes['zip'].dtype)
print(fines_by_zip['Zip Code'].dtype)
```

```python
zipcodes['zip'] = zipcodes['zip'].astype(str).str.zfill(5)  # ensures '00501' not '501'
fines_by_zip['Zip Code'] = fines_by_zip['Zip Code'].astype(str).str.zfill(5)
```

```python
zips_with_fines = pd.merge(left = zipcodes, right = fines_by_zip, 
                    left_on = 'zip', right_on = 'Zip Code', how='left')
zips_with_fines
```

```python
zips_with_fines = zips_with_fines.drop(columns = 'Zip Code')
```

```python
zips_with_fines = zips_with_fines.fillna(0)
```

```python
zips_with_fines['Missed Count'] = zips_with_fines['Missed Count'].astype('Int64')
```

```python
zips_with_fines['Fines'] = zips_with_fines['Fines'].astype('Int64')
```

```python
zips_with_fines
```

```python
fig, ax = plt.subplots(figsize=(10,10))
zips_with_fines.plot(
    column = 'Fines',
    cmap = 'Reds', 
    legend = True,
    edgecolor = 'black', 
    ax = ax)
plt.title('Fine Totals by Zip Code', fontsize=14)
ax.axis('off');
plt.savefig('../visualizations/fines_map.png', bbox_inches='tight', pad_inches=0.1)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
zips_with_fines.plot(
    column = 'Missed Count',
    cmap = 'Greys', 
    legend = True,
    edgecolor = 'black', 
    ax = ax)
plt.title('Missed Trash Pickups by Zip Code', fontsize=14)
ax.axis('off');
plt.savefig('../visualizations/missed_trash_map.png', bbox_inches='tight', pad_inches=0.1)
```
